```python
import numpy as np
import pandas as pd

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data for the first graph
data1 = {
    "Num_Workers": [1, 1, 2, 2],
    "CPU": [1, 2, 1, 2],
    "Memory": [4096, 4096, 4096, 4096],
    "Average Time": ["00:00:36", "00:00:31", "00:01:15", "00:01:51"]
}

# Data for the second graph
data2 = {
    "Num_Workers": [1, 1, 2, 2],
    "CPU": [1, 2, 1, 2],
    "Memory": [2048, 2048, 2048, 2048],
    "Average Time": ["00:08:33", "00:07:00", "00:02:17", "00:00:31"]
}

# Converting to DataFrame
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Function to convert time strings to seconds
def time_to_seconds(time_str):
    h, m, s = map(int, time_str.split(':'))
    return h * 3600 + m * 60 + s

# Convert Average Time to seconds for both dataframes
df1['Average Time'] = df1['Average Time'].apply(time_to_seconds)
df2['Average Time'] = df2['Average Time'].apply(time_to_seconds)

# Creating subplots for the two graphs
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Plotting for first graph (Memory 4096)
sns.lineplot(ax=axes[0], data=df1, x="Num_Workers", y="Average Time", hue="CPU", style="Memory", markers=True, dashes=False)
axes[0].set_title('Average Time vs Workers, CPU (Memory: 4096)')
axes[0].set_xlabel('Number of Workers')
axes[0].set_ylabel('Average Time (seconds)')

# Plotting for second graph (Memory 2048)
sns.lineplot(ax=axes[1], data=df2, x="Num_Workers", y="Average Time", hue="CPU", style="Memory", markers=True, dashes=False)
axes[1].set_title('Average Time vs Workers, CPU (Memory: 2048)')
axes[1].set_xlabel('Number of Workers')
axes[1].set_ylabel('Average Time (seconds)')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
